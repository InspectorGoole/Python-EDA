Now that we have cleaned our data of null values. It is time to deal with the outliers

```python
# create a new dataframe for big trees
# 245 rows of outliers

big_trees = tree_cen_subset[(tree_cen_subset['tree_dbh']>60) | (tree_cen_subset['stump_diam']>60)]
big_trees
```

1) One way to deal with the outliers would be to remove them completely
2) Another way is to find the type of trees they are and find the prefered statistical value for each type of tree, and replace the value

```python
tree_cen_subset = tree_cen_subset[(tree_cen_subset['tree_dbh']<=60) & (tree_cen_subset['stump_diam']<=60)]
tree_cen_subset
```

Now you have come across outliers that are too small, like having 0 diameter. 
You have 2 options 
1) remove the values that are zero, and the outliers that are super high
2) for each tree only except the values in the IQR, reject all values that are less than 25% and more than 75%

```python
stats_alive = tree_cen_subset_alive.groupby('spc_latin')['tree_dbh'].describe().reset_index()[['spc_latin', '25%', '75%']]
stats_alive
```

```python
tree_cen_subset_alive = tree_cen_subset_alive.merge(stats_alive, on='spc_latin', how='left')
tree_cen_subset_alive
```

The code checks where tree_dbh is greater than the value in the '75%' column for each row.
For those rows, it replaces the value in the tree_dbh column with the value from the '75%' column.

```python
mask = tree_cen_subset_alive['tree_dbh']<tree_cen_subset_alive['25%'] 
tree_cen_subset_alive.loc[mask, 'tree_dbh'] = tree_cen_subset_alive['25%']


mask = tree_cen_subset_alive['tree_dbh']>tree_cen_subset_alive['75%']
tree_cen_subset_alive.loc[mask, 'tree_dbh'] = tree_cen_subset_alive['75%']

# tree_cen_subset_alive['tree_dbh'] > tree_cen_subset_alive['75%'] creates a boolean condition that checks where values in the tree_dbh column are greater than the values in the '75%' column.
# mask is a boolean Series that stores True for rows where tree_dbh is greater than '75%' and False otherwise.

# tree_cen_subset_alive.loc[mask, 'tree_dbh'] selects the rows where the mask is True (i.e., where tree_dbh is greater than '75%').
# tree_cen_subset_alive['75%'] assigns the corresponding value from the '75%' column to the tree_dbh column for those selected rows.

# we have set the lower quartile as lower limit and upper quartile as upper limit
```
